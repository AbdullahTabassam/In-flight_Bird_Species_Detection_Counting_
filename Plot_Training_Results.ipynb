### Training loss vs Validation loss

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# Read the CSV file into a DataFrame
df = pd.read_csv('TUNED_lr0_0.01_lrf_0.001_warmup_epochs_5.0_warmup_momentum_0.95_optimizer_auto/train/results.csv', header=0)

# Extract the required columns
epochs = df.iloc[:,[0]]
train_loss = df.iloc[:,[3]]
val_loss = df.iloc[:,[10]]

# Plot the curves with different colors
plt.plot(epochs, train_loss, color='orange', label='Train Loss')
plt.plot(epochs, val_loss, color='tab:blue', label='Validation Loss')

# Set plot title and labels
plt.title('Loss Curves')
plt.xlabel('Epochs')
plt.ylabel('Loss')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

### mAP @ 0.5:0.05:0.95 IOU:

```python
map_50_95 = df.iloc[:,[7]]
plt.plot(epochs, map_50_95)

plt.title('mAP 50-95')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```
